## 섬의 개수
- 지도의 너비 W와 H가 주어짐
- 1은 땅 바다는 0

```python

W_list, H_list, maps_list = [], [], []
while True:
    W, H = map(int, input().split())
    if W==0 and H==0:
        break
    W_list.append(W)
    H_list.append(H)
    
    maps = []
    for _ in range(H):
        maps.append(list(map(int, input().split())))
    maps_list.append(maps)

from collections import deque
# 좌 우 상 하 좌상 좌하 우상 우하
dx = [-1, 1, 0, 0, -1, -1 ,1, 1]
dy = [0, 0, -1, 1, -1, 1, -1, 1]


def bfs(y, x, maps, H, W):
    queue = deque()
    queue.append((y,x))
    while queue:
        y, x = queue.popleft()
        maps[y][x] = 0
        for k in range(8):
            ny = y+dy[k]
            nx = x+dx[k]
            if ny<0 or nx<0 or ny>=H or nx>=W:
                continue
            if maps[ny][nx] == 0:
                continue
            queue.append((ny,nx))
            maps[ny][nx]=0
    
cnt_list = []
for W, H, maps in zip(W_list, H_list, maps_list):
    count = 0
    for i in range(H):
        for j in range(W):
            if maps[i][j] == 1:
                bfs(i, j, maps, H, W)
                count+=1
    cnt_list.append(count)

    
for m in cnt_list:
    print(m)
    
```
